```python
import plotly.offline as py
import plotly.graph_objs as go
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.ticker as mtick
import numpy as np



df = pd.read_excel("C:/Users/user/Documents/Datos para graficos python/EDAD.xlsx")
df= df[["SEXO 2019", "EDAD 2019"]].dropna()




bins = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 85, np.inf]

redad = ["21-25", "26-30", "31-35", "36-40", "41-45", "46-50", "51-55", "56-60", "61-65", "66-70", "71-75", "76-80",
        "81-85", "+86"]
 

df["Rango EDAD"] = pd.cut(df['EDAD 2019'], bins, labels = redad)

df = df.groupby(["SEXO 2019", "Rango EDAD"]).size().reset_index(name="cantidad")
dff = df.loc[df["SEXO 2019"] == "Femenino"]      
dfm = df.loc[df["SEXO 2019"] == "Masculino"]
dff["cantidad"] = (dff["cantidad"]*[-1])



women_bins = np.array(dff["cantidad"].tolist())
men_bins = np.array(dfm["cantidad"].tolist())

y = redad

layout = go.Layout(yaxis=go.layout.YAxis(title='Edad'),
                   xaxis=go.layout.XAxis(
                       range=[-240, 240],
                       tickvals=[-240, -120, -60, 0, 60, 120, 240],
                       ticktext=[240, 120, 60, 0, 60, 120, 240],
                       title='Cantidad de miembros'),
                       legend=dict(
                                x=0,
                                y=1,
                                traceorder='normal',
                                font=dict(
                                size=12,)),
                   barmode='overlay',
                   bargap=0.1, 
                   paper_bgcolor='rgba(0,0,0,0)',
                   plot_bgcolor='rgba(0,0,0,0)',
                            
                  )

data = [go.Bar(y=y,
               x=men_bins,
               orientation='h',
               name='Hombres',
               text= men_bins.astype('int'),
               textposition = "outside",
               marker=dict(color='rgba(255, 192 , 0, 1)')
               ),
        go.Bar(y=y,
               x=women_bins,
               orientation='h',
               name='Mujeres',
               text=-1 * women_bins.astype('int'),
               textposition = "outside",
               marker=dict(color='rgba(0, 176 , 80, 1)')
               )]
               
               
py.plot(dict(data=data, layout=layout))



```
